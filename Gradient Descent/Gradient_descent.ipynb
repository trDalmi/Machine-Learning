```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples = 4,n_features = 1, n_informative = 1,n_targets=1,noise = 80 , random_state = 13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
# Applying OLS(ordinary least sqaure)
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color = 'red')
```

```python
# Let's apply gradient descent assuming slope is constant m = 78.35
# and let's assume the starting value of intercept b = 0
y_pred = ((78.35*X) + 0).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color = 'blue',label = 'OLS')
plt.plot(X,y_pred,color = 'red',label = 'b=0')
plt.legend()
plt.show()
```

```python
# Iteration 1
m = 78.35
b = 0

loss_slope = -2*np.sum(y - m*X.ravel()-b)
loss_slope
```

```python
lr = 0.1
step_size = lr * loss_slope
step_size
```

```python
b1 = b - step_size
b1
```

```python
y_pred1 = ((78.35 *X) +b1).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color = 'blue',label = 'OLS')
plt.plot(X,y_pred,color = 'red',label = 'b=0')
plt.plot(X,y_pred1,color = 'orange', label = 'b={}'.format(b1))
plt.legend()
plt.show()
```

```python
# Iteration 2
loss_slope = -2 * np.sum(y - m*X.ravel()-b1)
loss_slope
```

```python
step_size  = loss_slope*lr
step_size
```

```python
b2 = b1 - step_size
b2
```

```python
y_pred2 = ((78.35 *X) +b2).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color = 'blue',label = 'OLS')
plt.plot(X,y_pred,color = 'red',label = 'b = 0')
plt.plot(X,y_pred1,color = 'orange', label = 'b={}'.format(b1))
plt.plot(X,y_pred2,color = 'green', label = 'b={}'.format(b2))
plt.legend()
plt.show()
```

```python
#Iteration 3
loss_slope = -2 * np.sum(y - m*X.ravel() - b2)
loss_slope
```

```python
step_size = lr*loss_slope
step_size
```

```python
b3  = b2 - step_size
b3
```

```python
y_pred3 = ((78.35*X) + b3).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color = 'red',label = 'OLS')
plt.plot(X,y_pred1,color = 'orange', label = 'b={}'.format(b1))
plt.plot(X,y_pred2,color = 'green', label = 'b={}'.format(b2))
plt.plot(X,y_pred3,color = 'black',label = 'b = {}'.format(b3))
```

```python
b = -100
m = 78.35
lr = 0.01

epochs = 100
for i in range(epochs):
  loss_slope = -2 * np.sum(y-m*X.ravel()-b)
  b = b - (lr * loss_slope)
  y_pred = m * X + b
  plt.plot(X,y_pred)
plt.scatter(X,y)
```
